# To Display All The Header Tags From Wikipedia.org# 

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

heading=[]
head1=[]


result=[]


page = requests.get("https://www.wikipedia.org/")

soup=BeautifulSoup(page.content)

for  element in soup.find_all("h1", class_="central-textlogo-wrapper"):
    heading.append(element.text.replace("\n",' '))
    
for element1 in soup.find_all("h2",class_="bookshelf-container"):
    head1.append(element1.text.replace("\n"," "))

result=heading+head1
#print(len(heading),len(head1))

df=pd.DataFrame({"H1 Tag":heading})
print(df)

print()  #H1 And H2 Len Is Not Same That's Y The Frame Is Saprated Print 

df1=pd.DataFrame({"H2 Tag":head1})

print(df1)
```

# Display IMDB’s Top Rated 100 Movies’ Data ( Name, Rating, Year Of Release)


```python
from bs4 import BeautifulSoup
import requests
import pandas as pd


page1=requests.get("https://www.imdb.com/chart/top/?ref_=nv_mv_250")

soup1=BeautifulSoup(page1.content)

title=[]

date1=[]

rating=[]

for ele in soup1.find_all('td',class_='titleColumn'):
    title.append(ele.text.splitlines()[2])

 

    
for ele2 in soup1.find_all('span',class_="secondaryInfo"):
    a=ele2.text.replace('('," ")
    date1.append(a.split(')')[0])
    

    
for ele3 in soup1.find_all('td',class_='ratingColumn imdbRating'):
    str1=str(ele3.text)
    li=(str1.splitlines()[1])
    rating.append(li)

    

df=pd.DataFrame({'Moive name':title,'Release Year':date1,'Rateing':rating})


df.head(100)
```

# Python Program To Scrape Mentioned Details From(dineout.co.in)

```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd


name=[]

cuisine=[]

loc=[]

rating1=[]

imeges=[]


page2=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')

soup3=BeautifulSoup(page2.content)

for ele11 in soup3.find_all('div',class_='restnt-info cursor'):
    name.append(ele11.text)
    
    
for ele13 in soup3.find_all('span',class_='double-line-ellipsis'):
    cuisine.append(ele13.text.split("|")[1])

    
for  ele12 in soup3.find_all('div',class_="restnt-loc ellipsis"):
    loc.append(ele12.text)
    

for ele14 in soup3.find_all('div',class_='restnt-rating rating-4'):
    rating1.append(ele14.text)
    

for ele15 in soup3.find_all('img',class_='no-img'):
    imeges.append(ele15['data-src'])

df4=pd.DataFrame({'Restaurant name':name,'Cuisine':cuisine,' Location':loc,'Ratings':rating1,'Image URL':imeges})


df4
```

# Scrape Weather Details For Last 24 Hours From (tutiempo.net)

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd


page=requests.get('https://en.tutiempo.net/delhi.html?data=last-24-hours')

soup=BeautifulSoup(page.content)


humdity=[]

t1=[]

time=[]

wind=[]

temp=[]

press=[]

f1=[]



    
for i in soup.find_all('td',class_='hr'):
    humdity.append(i.text)
    
for i8 in soup.find_all('tr')[5::2]:
    t1.append(i8.text[0:5])
t1.remove('Yeste')
time.extend(t1)
    
for i3 in soup.find_all('td',class_='wind'):
    wind.append(i3.text)
    
for i4 in soup.find_all('td',class_='t Temp'):
    temp.append(i4.text)
    
for i2 in soup.find_all('td',class_='prob'):
    press.append(i2.text)
    
for i5 in soup.find_all('span',class_='thhip ico i0530 u3018'):
    f1.append(i5.text)
for i6 in soup.find_all('span',class_='thhip ico i0530 u303'):
    f1.extend(i6.text.splitlines())
for i7 in soup.find_all('span',class_='thhip ico i0530 u303n'):
    f1.extend(i7.text.splitlines())

    
df1=pd.DataFrame({'Hour':time,'Temperature':temp,'Wind':wind,'Weather condition':f1,'Humidity':humdity,'Pressure':press})

df1
```


# Python Program To Scrape Monument Name, Monument Description, Image URL About Top 10 Monuments From (puredestinations.co.uk)

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

page=requests.get('https://www.puredestinations.co.uk/top-10-famous-monuments-to-visit-in-india/')

soup=BeautifulSoup(page.content)


mname=[]

mtext=[]

imges=[]



for i in soup.find_all('strong')[:10]:
    mname.append(i.text)
    
    
for i1 in soup.find_all('p')[5:35:3]:
    mtext.append(i1.text)
    

for i2 in soup('img')[6:26:2]:
    imges.append(i2['src'])
    
    
df=pd.DataFrame({'Monument Name':mname,'Monument Description':mtext,'Image URL':imges})


df
```

# a) Top 10 ODI Teams In Men’s Cricket Along With The Records For Matches, Points And Rating.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

page=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')

soup=BeautifulSoup(page.content)


team=[]

math=[]

points=[]

rating=[]


nzm=soup.find('td',class_='rankings-block__banner--matches')
math.append(nzm.text)


pio=soup.find('td',class_='rankings-block__banner--points')
points.append(pio.text)


rat=soup.find('td',class_='rankings-block__banner--rating u-text-right')
rating.append(rat.text[29:].split('\n')[0])



for i in soup.find_all('span',class_='u-hide-phablet')[:20]:
    team.append(i.text)
    
    
for i2 in soup.find_all('td',class_='table-body__cell u-center-text')[::2]:
    math.extend(i2.text.splitlines())

    
for i3 in soup.find_all('td',class_='table-body__cell u-center-text')[1::2]:
    points.extend(i3.text.splitlines())
    
    
for i4 in soup.find_all('td',class_='table-body__cell u-text-right rating'):
    rating.extend(i4.text.split())

    
df=pd.DataFrame({'Team Name':team,'Matches':math,' Points':points,'Rating':rating})



df.head(10)
```

# To Scrape Details Of All The Posts From coreyms.com. Scrape The Heading, Date, content And The Code For The Video From The Link For The Youtube Video From The Post.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd


page=requests.get('https://coreyms.com/')

soup=BeautifulSoup(page.content)

links1=[]

date=[]

content1=[]

heading=[]


for i in soup.find_all('a',class_='entry-title-link'):
    links1.append(i['href'])
    
    
for i1 in soup.find_all('time',class_='entry-time'):
    date.append(i1.text)
    
    
for i2 in soup.find_all('div',class_='entry-content'):
    content1.append(i2.text.replace('\n',''))
    
    
    
for i3 in soup.find_all('h2',class_='entry-title'):
    heading.append(i3.text)
    

df=pd.DataFrame({'Headding':heading,'Date':date,'Contents':content1,'YouTube Link':links1})


df
```
